```python
import tgan
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('/content/drive/MyDrive/PAPER/진행연구/tabular deep learning/폐암병기/data/11.train_data_73_death_smoker.csv')

continuous_columns = [4,5,9,10,11]
```

```python
'''
                        'a_year_time',
                        'a_year_event',
                        'age',
                        'Smoker',
                        'PackYear',
                        'ECOG',
                        'FVC',
                        'FEV1',
                        'DLCO',
                        'DLCO_PERCENT',
                        'SmallCellC',
                        'AdenoC',
                        'largeCellC',
                        'Positive_EGFR_Mutation',
                        'Positive_ALK_IHC',
                        'Positive_ALK_FISH',
                        'STAGE_simple',
                        'OP_Curative',
                        'RT',
                        'Chemo_curative',
                        'Chemo_Palliative'
'''
```

```python
from tgan.model import TGANModel
tgan = TGANModel(continuous_columns)
```

```python
tgan = TGANModel(
    continuous_columns,
    output='output',
    gpu=None,
    max_epoch=5,
    steps_per_epoch=10000,
    save_checkpoints=True,
    restore_session=True,
    batch_size=100,
    z_dim=200,
    noise=0.2,
    l2norm=0.00001,
    learning_rate=0.001,
    num_gen_rnn=100,
    num_gen_feature=100,
    num_dis_layers=1,
    num_dis_hidden=100,
    optimizer='AdamOptimizer'
)
```

```python
tgan.fit(data)
```

```python
num_samples = 15600
                                                            
samples = tgan.sample(num_samples)

samples.head(3)
```
